```python
"""
author : pavan pyla 
reg no: 22232
"""
"""Q9. Write a function n_odds(x),
 where  𝑥  is a one-dimensional array of floats and the output is the number of odd numbers in  𝑥."""


def n_odds(x):
    """this function takes a list as argument 
    where  𝑥  is a one-dimensional array of floats and the output is the number of odd numbers in  𝑥.
    Args:
        x (_list_): _its contains some ramdom numbers in order_
    """
    count=0
    for i in range(len(x)):            #iterating through the list to get all the odd numbers
      if(x[i] % 2.0 !=0):
         count=count+1

    return count                       #returning the no of odd numbers in list

x=[1,2.0,3,45,6,7,8,9.0,90,0,3,4,5,6,7,8.0,9]           #the list
count=n_odds(x)
print(f"the no of odd number in the array is :  {count}")   
```

```python
"""
author : pavan pyla 
reg no: 22232
"""
"""Q10. Write a function vol_cylinder(r,h), 
where  𝑟  and  ℎ  are the radius and height of a cylinder, respectively, 
and the output is a list  [𝑠,𝑣]  where  𝑠  and  𝑣  are the surface area and volume of the same cylinder,
 respectively. Recall that the surface area of a cylinder is  2π𝑟2+2π𝑟ℎ,  and the volume is  π𝑟2ℎ. 
  Assume that  𝑟  and  ℎ  are 1 by 1 float.
"""

from math import pi


def vol_cylinder(r,h):
   """_this function takes 2 floats as orguments (radius and height of a cylinder)_
   and the output is a list  [𝑠,𝑣]  where  𝑠  and  𝑣  are the surface area and volume of the same cylinder,
 respectively. Recall that the surface area of a cylinder is  2π𝑟2+2π𝑟ℎ,  and the volume is  π𝑟2ℎ.

   Args:
       r (_float_): _radius of the given cylinder_
       h (_float_): _height of the given cylinder_
   """
   v=pi*(r**2)*h
   s=2*pi*r*(r+h)                   #calculating the both voume and surface area and returning them as list
   return [v,s]

r=int(input(" enter the radius of the cylinder  : "))            #taking inputs from user
h=int(input(" enter the height of the cylinder : "))
result=vol_cylinder(r,h)
print(f"the volume of the cylinder {result[0]} and the surface area of cylinder is {result[1]}")    #printing the results
```
