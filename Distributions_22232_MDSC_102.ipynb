## Binomial Distribution 

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

from math import factorial
                                    #importing the required functions
from numpy import power




def binomial_fun(r,n,p):
   if(r>n):
      print("\n '0' as x is grater than no of trails   ")
      return
   if(p>1 or p<0):
      print(" \n the probability can not be more than 1 or less than 0 \n ")
      return
   if(n<0):
      print("no of trails can not be less than '0' ")
      return
   
   ncr= factorial(n) / (factorial(r) * factorial(n-r))
   prob= ncr*(power(p,r))*(power(1-p,n-r))
   return prob





def plot_binom(n1=30,p1=0.5):
    """_this function takes 2 orguments n1 that is no of trails and p1 that is probability of success.
    Then prints the graph of binomial distribution using those parameters _

    Args:
        n1 (int, optional): _no of trails and the constraint is n>0 and it should be an integre_. Defaults to 30.
        p1 (float, optional): _probability of success and the constraint is p1>0_. Defaults to 0.5.
    """
    x=[]
    n=n1
    p=p1
    for i in range(n+1):
        x.append(i)
    l=[]
    for i in range(n+1):
        l.append(binomial_fun(i,n,p))
    plt.bar(x,l,color="red")
    plt.plot(x,l,color="g")
    plt.xlabel("Value of x in NcX",fontsize="18")
    plt.ylabel("Probability",fontsize="18")
```

```python
n=int(input("enter the no of trails : "))
p=float(input("enter the probability of succuess : "))
plot_binom(n,p)
```

## Normal Distribution 

```python
def f(x,u,s):
    return ( 1/(np.sqrt(2*np.pi*s)) )*( np.exp(  (-1/2)*((x-u)/s )**2)  )

def plot_normal(u1=0,s1=1):
    """_this function takes 2 orguments u1 that is mean of normal distribution  and s1 as sigma (varience )of the normal distribution 
    and using those values the function will plot the graph of Normal Distribution_

    Args:
        u1 (float, optional): _mean of the distribution _. Defaults to 0.
        s1 (float, optional): _varience of the distribution ,can not be < 0_. Defaults to 1.
    """
    u=u1
    s=s1
    x=np.linspace(u-3*s,u+3*s,1000)
    plt.figure(figsize = (12,7))
    plt.plot(x,f(x,u,s),color="red")
    plt.bar(x,f(x,u,s),color="b")
```

```python
plot_normal(5,7)
```

## Geometric Distribution

```python
import numpy as np
import math
import matplotlib.pyplot as plt


def Geometric(n,p,x):
    return p*(1-p)**(x-1)
def plot_Geo(n,p):
    """_this function takes 2 orguments n,p that is n is nth trail and p is probability of succuess.
    and then it prints the graph of the Geometric distribution_

    Args:
        n (_int_): _nth trail succuess_
        p (_float_): _probability of succuesss_
    """
    
    X = [i for i in range(n+1)]
    y = []
    for x in X:
        y.append(Geometric(n,p,x))
        
    plt.title(f'Plot of Geometric Random Variable from {X[0]} to {X[-1]}')
    plt.xlabel('x', fontsize = 18)
    plt.ylabel('f(x)', fontsize = 18)   
    plt.plot(X,y,'g-')
    #plt.bar(X,y,color='r')

    plt.show()
p = float(input("Enter the probablity of success :"))
n = int(input("Enter the value of N :"))
plot_Geo(n,p)
```

## Gamma Distribution

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate as integrate

def gammarv(alpha,value,l,x):                              #computing the gamma Rv
    return (l**alpha)/value * x**(alpha-1) * np.exp(-l*x)


def GammaInt(alpha,x):                          #computing the gamma integral
    return x**(alpha-1) * np.exp(-x)



def Plot_Gamma(a,l):
    """_this functions takes 2 orguments 1 is alpha and 2 is lamda and
     prints the graph of the Gamma distribution using those parameters_

    Args:
        a (_float_): _value of alpha ,should be >0_
        l (_flaot_): _value of lamda ,should be >0_
    """



    X = np.linspace(0.0000001,10,500)
    y = []
    for x in X:
        value=GammaInt(a,x)
        y.append(gammarv(a,value,l,x))
        
    plt.title(f' Gamma Distribution ')
    plt.xlabel('X', fontsize = 18)
    plt.ylabel('probality', fontsize = 18)   
    plt.plot(X,y,'r')
    #plt.bar(X,y,color="yellow")
    plt.show()


l = float(input("Enter lambda value :"))
a = float(input("Enter alpha value :"))
Plot_Gamma(a,l)
```
